```python
import pandas as pd
```

```python
"""
Import all pre-cleaned data
Data here was sourced from ~10 spreadsheets per year
Not done with a function because column names differ from year to year
"""

# import 2011-2012 data
df2012 = pd.read_csv('./CSVfiles/book2012.csv')
sf2012 = pd.read_csv('./CSVfiles/seniors2012.csv')
ff2012 = pd.read_csv('./CSVfiles/financial2012.csv')

# combine first and last names
df2012['total_name']= df2012['First Name'].str.replace(' ','').replace('-','') \
                    + df2012['Last Name'].str.replace(' ','').replace('-','')
sf2012['total_name']= sf2012['Name'].str.replace(' ','').replace('-','')
ff2012['total_name']= ff2012['First'].str.replace(' ','').replace('-','') \
                    + ff2012['Last'].str.replace(' ','').replace('-','')

# import 2012-2013 data
df2013 = pd.read_csv('./CSVfiles/book2013.csv')
sf2013 = pd.read_csv('./CSVfiles/seniors2013.csv')
ff2013 = pd.read_csv('./CSVfiles/financial2013.csv')

# combine first and last names
df2013['total_name']= df2013['First Name'].str.replace(' ','').replace('-','') \
                    + df2013['Last Name'].str.replace(' ','').replace('-','')
sf2013['total_name']= sf2013['Name'].str.replace(' ','').replace('-','')
ff2013['total_name']= ff2013['First'].str.replace(' ','').replace('-','') \
                    + ff2013['Last'].str.replace(' ','').replace('-','')

# import 2013-2014 data
df2014 = pd.read_csv('./CSVfiles/book2014.csv')
sf2014 = pd.read_csv('./CSVfiles/seniors2014.csv')
ff2014 = pd.read_csv('./CSVfiles/financial2014.csv')

# combine first and last names
df2014['total_name']= df2014['Full Name (First Name)'].str.replace(' ','').replace('-','') \
                    + df2014['Full Name (Last Name)'].str.replace(' ','').replace('-','')
sf2014['total_name']= sf2014['First Name'].str.replace(' ','').replace('-','') \
                    + sf2014['Last Name'].str.replace(' ','').replace('-','')
ff2014['total_name']= ff2014['First'].str.replace(' ','').replace('-','') \
                    + ff2014['Last'].str.replace(' ','').replace('-','')

# import 2014-2015 data
df2015 = pd.read_csv('./CSVfiles/book2015.csv')
sf2015 = pd.read_csv('./CSVfiles/seniors2015.csv')
ff2015 = pd.read_csv('./CSVfiles/financial2015.csv')

# combine first and last names
df2015['total_name']= df2015['Full Name (First Name)'].str.replace(' ','').replace('-','') \
                    + df2015['Full Name (Last Name)'].str.replace(' ','').replace('-','')
sf2015['total_name']= sf2015['Student First Name'].str.replace(' ','').replace('-','') \
                    + sf2015['Student Last Name'].str.replace(' ','').replace('-','')
ff2015['total_name']= ff2015['First'].str.replace(' ','').replace('-','') \
                    + ff2015['Last'].str.replace(' ','').replace('-','')

# import 2015-2016 data
df2016 = pd.read_csv('./CSVfiles/book2016.csv')
sf2016 = pd.read_csv('./CSVfiles/seniors2016.csv')
ff2016 = pd.read_csv('./CSVfiles/financial2016.csv')

# combine first and last names
df2016['total_name']= df2016['Full Name (First Name)'].str.replace(' ','').replace('-','') \
                    + df2016['Full Name (Last Name)'].str.replace(' ','').replace('-','')
sf2016['total_name']= sf2016['Student First Name'].str.replace(' ','').replace('-','') \
                    + sf2016['Student Last Name'].str.replace(' ','').replace('-','')
ff2016['total_name']= ff2016['First'].str.replace(' ','').replace('-','') \
                    + ff2016['Last'].str.replace(' ','').replace('-','')

# import 2016-2017 data
df2017 = pd.read_csv('./CSVfiles/book2017.csv')
sf2017 = pd.read_csv('./CSVfiles/seniors2017.csv')
ff2017 = pd.read_csv('./CSVfiles/financial2017.csv')

# combine first and last names
df2017['total_name']= df2017['First Name'].str.replace(' ','').replace('-','') \
                    + df2017['Last Name'].str.replace(' ','').replace('-','')
sf2017['total_name']= sf2017['Student First Name'].str.replace(' ','').replace('-','') \
                    + sf2017['Student Last Name'].str.replace(' ','').replace('-','')
ff2017['total_name']= ff2017['First'].str.replace(' ','').replace('-','') \
                    + ff2017['Last'].str.replace(' ','').replace('-','')

# import 2017-2018 data
df2018 = pd.read_csv('./CSVfiles/book2018.csv')
sf2018 = pd.read_csv('./CSVfiles/seniors2018.csv')
ff2018 = pd.read_csv('./CSVfiles/financial2018.csv')

# combine first and last names
df2018['total_name']= df2018['First Name'].str.replace(' ','').replace('-','') \
                    + df2018['Last Name'].str.replace(' ','').replace('-','')
sf2018['total_name']= sf2018['First Name'].str.replace(' ','').replace('-','') \
                    + sf2018['Last Name'].str.replace(' ','').replace('-','')
ff2018['total_name']= ff2018['First Name'].str.replace(' ','').replace('-','') \
                    + ff2018['Last Name'].str.replace(' ','').replace('-','')

# drop crazy spurious empty columns
df2012.dropna(how = 'all', axis=1, inplace=True)
df2013.dropna(how = 'all', axis=1, inplace=True)
df2014.dropna(how = 'all', axis=1, inplace=True)
df2015.dropna(how = 'all', axis=1, inplace=True)
df2016.dropna(how = 'all', axis=1, inplace=True)
df2017.dropna(how = 'all', axis=1, inplace=True)
df2018.dropna(how = 'all', axis=1, inplace=True)

"""
merge data into temporary structures
"""
# 2012 data
temp2012 = pd.merge(sf2012,df2012, on = 'total_name')
temp2012 = pd.merge(temp2012,ff2012, on = 'total_name', how = 'left')

# 2013 data
temp2013 = pd.merge(sf2013,df2013, on = 'total_name')
temp2013 = pd.merge(temp2013,ff2013, on = 'total_name', how = 'left')

# 2014 data
temp2014 = pd.merge(sf2014,df2014, on = 'total_name')
temp2014 = pd.merge(temp2014,ff2014, on = 'total_name', how = 'left')

# 2015 data
temp2015 = pd.merge(sf2015,df2015, on = 'total_name')
temp2015 = pd.merge(temp2015,ff2015, on = 'total_name', how = 'left')

# 2016 data
temp2016 = pd.merge(sf2016,df2016, on = 'total_name')
temp2016 = pd.merge(temp2016,ff2016, on = 'total_name', how = 'left')

# 2017 data
temp2017 = pd.merge(sf2017,df2017, on = 'total_name')
temp2017 = pd.merge(temp2017,ff2017, on = 'total_name', how = 'left')

# 2018 data
temp2018 = pd.merge(sf2018,df2018, on = 'total_name')
temp2018 = pd.merge(temp2018,ff2018, on = 'total_name', how = 'left')

"""
Save merged files"""
temp2012.to_csv('./MergedCSV/2012.csv')
temp2013.to_csv('./MergedCSV/2013.csv')
temp2014.to_csv('./MergedCSV/2014.csv')
temp2015.to_csv('./MergedCSV/2015.csv')
temp2016.to_csv('./MergedCSV/2016.csv')
temp2017.to_csv('./MergedCSV/2017.csv')
temp2018.to_csv('./MergedCSV/2018.csv')


# """
# Print out size of structures to check shapes
# """
# print('2012: ', df2012.shape, ff2012.shape, sf2012.shape, temp2012.shape)
# print('2013: ', df2013.shape, ff2013.shape, sf2013.shape, temp2013.shape)
# print('2014: ', df2014.shape, ff2014.shape, sf2014.shape, temp2014.shape)
# print('2015: ', df2015.shape, ff2015.shape, sf2015.shape, temp2015.shape)
# print('2016: ', df2016.shape, ff2016.shape, sf2016.shape, temp2016.shape)
# print('2017: ', df2017.shape, ff2017.shape, sf2017.shape, temp2017.shape)
# print('2018: ', df2018.shape, ff2018.shape, sf2018.shape, temp2018.shape)

"""
Set up new dataframes for consistent columns
"""
data2012 = pd.DataFrame()
data2013 = pd.DataFrame()
data2014 = pd.DataFrame()
data2015 = pd.DataFrame()
data2016 = pd.DataFrame()
data2017 = pd.DataFrame()
data2018 = pd.DataFrame()

"""
Insert Values
"""

# insert names
data2012['name'] = temp2012['total_name'].values
data2013['name'] = temp2013['total_name'].values
data2014['name'] = temp2014['total_name'].values
data2015['name'] = temp2015['total_name'].values
data2016['name'] = temp2016['total_name'].values
data2017['name'] = temp2017['total_name'].values
data2018['name'] = temp2018['total_name'].values

# insert universities
data2012['uni']=temp2012['School_x'].values
data2013['uni']=temp2013['School'].values
data2014['uni']=temp2014['School'].values
data2015['uni']=temp2015['School you will be attending in Fall 2015'].values
data2016['uni']=temp2016['School you will be attending in Fall 2015'].values
data2017['uni']=temp2017['School you will be attending in Fall 2017'].values
data2018['uni']=temp2018['School you will be attending in Fall 2018'].values

# insert instruments
data2012['instrument']=temp2012['Instrument_x'].values
data2013['instrument']=temp2013['Instrument'].values
data2014['instrument']=temp2014['Instrument'].values
data2015['instrument']=temp2015['Instrument played during the 2014-2015 CYSO Season'].values
data2016['instrument']=temp2016['Instrument played during the 2014-2015 CYSO Season'].values
data2017['instrument']=temp2017['Instrument played during the 2016-2017 CYSO Season'].values
data2018['instrument']=temp2018['Instrument played during 2017-2018 season (can select more than one)'].values

# insert graduation year
data2012['gradyear']=2012
data2013['gradyear']=2013
data2014['gradyear']=2014
data2015['gradyear']=2015
data2016['gradyear']=2016
data2017['gradyear']=2017
data2018['gradyear']=2018

# insert college major
data2012['major']=temp2012['Major'].values
data2013['major']=temp2013['Major'].values
data2014['major']=temp2014['Major'].values
data2015['major']=temp2015['Major (If double majoring include both here)'].values
data2016['major']=temp2016['Major (If double majoring include both here)'].values
data2017['major']=temp2017['Major (If double majoring include both here)'].values
data2018['major']=temp2018['Major(s) and Minor(s)'].values

# insert financial aid values
data2012['finaid']=temp2012['Fin Value'].values
data2013['finaid']=temp2013['Fin Value'].values
data2014['finaid']=temp2014['Fin Value'].values
data2015['finaid']=temp2015['Fin Value'].values
data2016['finaid']=temp2016['Fin Score'].values
data2017['finaid']=temp2017['Fin Value'].values
data2018['finaid']=temp2018['Fin Rank'].values

# insert ethnicity
data2012['race']=temp2012['Race/Ethnicity'].values
data2013['race']=temp2013['Racial/Ethnic Background (check all that apply)'].values
data2014['race']=temp2014['Racial/Ethnic Background (check all that apply)'].values
data2015['race']=temp2015['Racial/Ethnic Background (check all that apply)'].values
data2016['race']=temp2016['Racial/Ethnic Background (check all that apply)'].values
data2017['race']=temp2017['Racial/Ethnic Background (check all that apply)'].values
data2018['race']=temp2018['Racial/Ethnic Background (check all that apply)'].values

# insert gender
data2012['gender']=temp2012['GENDER'].values
data2013['gender']=temp2013['Gender'].values
data2014['gender']=temp2014['Gender'].values
data2015['gender']=temp2015['Gender'].values
data2016['gender']=temp2016['Gender'].values
data2017['gender']=temp2017['Gender'].values
data2018['gender']=temp2018['Gender'].values

# mother's employer
data2012['m_employ']=temp2012["Mother's Employer"].values
data2013['m_employ']=temp2013["Employer"].values
data2014['m_employ']=temp2014["Employer"].values
data2015['m_employ']=temp2015["Employer"].values
data2016['m_employ']=temp2016["Employer"].values
data2017['m_employ']=temp2017["Employer"].values
data2018['m_employ']=temp2018["Employer"].values

# mother's job title
data2012['m_job']=temp2012["Mother's Position"].values
data2013['m_job']=temp2013["Position"].values
data2014['m_job']=temp2014["Position"].values
data2015['m_job']=temp2015["Position"].values
data2016['m_job']=temp2016["Position"].values
data2017['m_job']=temp2017["Position"].values
data2018['m_job']=temp2018["Position"].values

# fathers's employer
data2012['d_employ']=temp2012["Father's Employer"].values
data2013['d_employ']=temp2013["Employer.1"].values
data2014['d_employ']=temp2014["Employer.1"].values
data2015['d_employ']=temp2015["Employer.1"].values
data2016['d_employ']=temp2016["Employer.1"].values
data2017['d_employ']=temp2017["Employer 2"].values
data2018['d_employ']=temp2018["Employer 2"].values

# fathers's job title
data2012['m_job']=temp2012["Father's Position"].values
data2013['m_job']=temp2013["Position.1"].values
data2014['m_job']=temp2014["Position.1"].values
data2015['m_job']=temp2015["Position.1"].values
data2016['m_job']=temp2016["Position.1"].values
data2017['m_job']=temp2017["Position 2"].values
data2018['m_job']=temp2018["Position 2"].values

# home zip code
data2012['homezip']=temp2012['Zip'].values
data2013['homezip']=temp2013['Postal / Zip Code'].values
data2014['homezip']=temp2014['Postal / Zip Code'].values
data2015['homezip']=temp2015['Postal / Zip Code'].values
data2016['homezip']=temp2016['Postal / Zip Code'].values
data2017['homezip']=temp2017['Postal / Zip Code'].values
data2018['homezip']=temp2018['Postal / Zip Code'].values

# highschool name
data2012['school']=temp2012['School_y'].values
data2013['school']=temp2013['School Name'].values
data2014['school']=temp2014['School Name'].values
data2015['school']=temp2015['School Name'].values
data2016['school']=temp2016['School Name'].values
data2017['school']=temp2017['School Name'].values
data2018['school']=temp2018['School Name'].values

# highschool address
data2012['s_address']=temp2012['School Address'].values
data2013['s_address']=temp2013['Street Address.3'].values
data2014['s_address']=temp2014['Street Address.3'].values
data2015['s_address']=temp2015['Street Address.3'].values
data2016['s_address']=temp2016['Street Address.3'].values
data2017['s_address']=temp2017['Street Address 4'].values
data2018['s_address']=temp2018['Street Address 3'].values

# highschool state
data2012['s_state']=temp2012['SCH State'].values
data2013['s_state']=temp2013['State / Province'].values
data2014['s_state']=temp2014['State / Province'].values
data2015['s_state']=temp2015['State / Province'].values
data2016['s_state']=temp2016['State / Province'].values
data2017['s_state']=temp2017['State / Province'].values
data2018['s_state']=temp2018['State / Province'].values

# highschool zip
data2012['s_zip']=temp2012['SCH Zip'].values
data2013['s_zip']=temp2013['Postal / Zip Code.3'].values
data2014['s_zip']=temp2014['Postal / Zip Code.3'].values
data2015['s_zip']=temp2015['Postal / Zip Code.3'].values
data2016['s_zip']=temp2016['Postal / Zip Code.3'].values
data2017['s_zip']=temp2017['Postal / Zip Code 4'].values
data2018['s_zip']=temp2018['Postal / Zip Code 3'].values

# school ensembles
data2012['s_ensemble']=temp2012['Band'].map(str) + temp2012['Orchestra'].map(str) + temp2012['Other'].map(str) +  temp2012['None'].map(str)
data2013['s_ensemble']=temp2013['Will you be playing in any of the following at your school during the 2012-13 season? (Please check all that apply)']
data2014['s_ensemble']=temp2014['Will you be playing in any of the following at your school during the 2013-14 season? (Please check all that apply)']
data2015['s_ensemble']=temp2015['Will you be playing in any of the following at your school during the 2013-14 season? (Please check all that apply)']
data2016['s_ensemble']=temp2016['Will you be playing in any of the following at your school during the 2015-16 season? (Please check all that apply)']
data2017['s_ensemble']=temp2017['Will you be playing in any of the following at your school during the 2016-17 season? (Please check all that apply)']
data2018['s_ensemble']=temp2018['Will you be playing in any of the following at your school during the 2017-18 season? (Please check all that apply)']

# private teacher name
data2012['pt']=temp2012['Private Teacher Name'].values
data2013['pt']=temp2013['First Name.3'] + temp2013['Last Name.3']
data2014['pt']=temp2014['Full Name (First Name).1'] + temp2014['Full Name (Last Name).1']
data2015['pt']=temp2015['Full Name (First Name).1'] + temp2015['Full Name (Last Name).1']
data2016['pt']=temp2016['Teacher Full Name (First Name)'] + temp2016[" Teacher Full Name (Last Name)"]
data2017['pt']=temp2017['Full Name (First Name)'] + temp2017["Full Name (Last Name)"]
data2018['pt']=temp2018['Full Name (First Name)'] + temp2018["Full Name (Last Name)"]

"""
Anonymize and Save Data for Further Processing
"""

student_data = pd.concat([data2012,data2013,data2014,data2015,data2016,data2017,data2018], ignore_index=True)
student_data.to_csv('./CYSOwithnames.csv')
student_data['studentID'] = student_data.index
student_data.drop(['name'],axis=1,inplace=True)
student_data.to_csv('/Users/tbowling/ds/metis/working/projects/CYSOadvancements/data/CYSOanonymized.csv')


```
